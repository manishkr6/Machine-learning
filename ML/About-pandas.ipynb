#### It randomly shuffles all the rows in the car_sales DataFrame.
##### .sample() is used to take a random sample of rows from a DataFrame.
##### frac=1 means you're taking 100% of the rows (i.e., the full DataFrame), but shuffled.

```python
car_sales_shuffled = car_sales.sample(frac=1)
```

```python
car_sales_shuffled
```

```python
# Only select 20% of data
car_sales_shuffled.sample(frac=0.01)
```

```python
# Re-set dataframe
car_sales_shuffled.reset_index(drop=True, inplace=True)
```
